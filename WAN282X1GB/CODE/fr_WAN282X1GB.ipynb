- 2) transfer pdf to docx

- 3) transfer pdf to txt

- 4) extract image(H1:image include a keyword text) for docx

- 6) save the keyword text = core text

```python
# Python 모듈용 Aspose.Words 가져오기
import aspose.words as aw

# PDF 파일을 로드하고 Word DOCX 형식으로 변환
pdf = aw.Document("user_manual.pdf")
pdf.save("user_manual.docx")
pdf.save("user_manual.txt")

# PDF의 DOCX 버전 로드
doc = aw.Document("user_manual.docx")

# 모든 모양 검색
shapes = doc.get_child_nodes(aw.NodeType.SHAPE, True)
imageIndex = 0

# 모양을 통해 루프
for shape in shapes :
    shape = shape.as_shape()
    if (shape.has_image) :

        # 이미지 파일 이름 설정
        imageFileName = f"manual_{imageIndex}{aw.FileFormatUtil.image_type_to_extension(shape.image_data.image_type)}"

        # 이미지를 저장
        shape.image_data.save(imageFileName)
        imageIndex += 1                                                                                                                                                  
```

```python
manuals = []

f = open('user_manual.txt','r', encoding = 'UTF-8')
user_manual = f.readlines()
for line in user_manual:
    manuals.append(line)
f.close()
```

```python
df = pd.DataFrame(manuals)
df.columns = ['body']
df.to_csv("user_manual_core.csv")
df
```

```python
import os
import docx2txt 

def docx_to_txt(path, filename):
    text = docx2txt.process(path + filename).split('\n') 
    # docx 파일의 내용이 text에 담기는데 '\n' 줄바꿈 문자 기준으로 나뉘어 리스트로 만들어 진다.
    newfile = os.path.splitext(filename)[0] + '.txt'
    f = open(path + newfile, 'w')    # txt 파일 오픈(없으면 생성됨)
    for item in text:    # text에 담긴 요소 하나가 txt파일에 한줄로 써지게 됨.
        f.write(item + '\n')                                
    f.close()
    os.remove(path + filename)        # {file].docx는 사용이 모두 끝났으므로 삭제
    return (newfile)                        # {file}.txt가 리턴됨
```

```python
import pickle
import re
import nltk
from nltk.corpus import stopwords
from tqdm import tqdm

component['clean_text'] = pd.Series(dtype='object')
component['stopwords_after'] = pd.Series(dtype='object')
component['pos_tag'] = pd.Series(dtype='object')
component['NN'] = pd.Series(dtype='object')
component['lemmatization'] = pd.Series(dtype='object')

stops = stopwords.words('english')
stops.append('machine')
stops.append('product')
stops.append('bosch')

def clean_text(texts): 
    corpus = []
    
    for i in tqdm(range(0, len(texts))):
        
        body = texts[i]
        
        body = re.sub('[^a-zA-Z]', ' ', body) # 특수문자 제거 
        body = body.lower().split() # 대문자를 소문자로 변경, 문장을 단어 단위로 구분
        
        component['clean_text'][i] = body
        
        no_stops = [word for word in body if not word in stops] # 불용어 제거
        component['stopwords_after'][i] = no_stops
        
        tokens_pos = nltk.pos_tag(component['stopwords_after'][i]) # pos tagging (품사 태깅)
        component['pos_tag'][i] = tokens_pos
        
        NN_words = [] # 명사는 NN을 포함하고 있음을 알 수 있음(명사만 추출)
        for word, pos in tokens_pos:
            if 'NN' in pos:
                NN_words.append(word)
                component['NN'][i] = NN_words
                
        wlem = nltk.WordNetLemmatizer() # Lemmatization(원형(lemma) 찾기) # nltk에서 제공되는 WordNetLemmatizer을 이용
        lemmatized_words = []
        for word in NN_words:
            new_word = wlem.lemmatize(word)
            lemmatized_words.append(new_word)
            component['lemmatization'][i] = lemmatized_words
        
        corpus.append(no_stops) 
        
    return corpus
```
