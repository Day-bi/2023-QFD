# 동시출현빈도 매트릭스를 만들기 위한 함수임. 여기선 사용 x

```python
def makedocx(df):
    docx = []
    for i in range(len(df)):
        doc = df['lemmatization'][i]
        docx.append(doc)
    return docx
```
