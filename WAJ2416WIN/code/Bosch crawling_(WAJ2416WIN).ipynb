## crawling part
- procedure a chromedriver
- define df:= title,star,date,contents & url --> forë¬¸
- click the next page
- generate df

```python
path = "C:/Users/nurip/Downloads"
driver = webdriver.Chrome()

#reviews = obj.select('div.total-review-count > span > div').get_text().strip()
titles = []
stars = []
dates = []
contents = []
    
url1 = "https://www.amazon.in/Bosch-Inverter-Control-Automatic-WAJ2416WIN/product-reviews/B08SR52BSM/ref=cm_cr_arp_d_paging_btm_next_2?ie=reviewerType=all_reviews&pageNumber="
url2 = 0
for url2 in range(1,218):
    url = url1 + str(url2)
    # driver get url
    driver.get(url)
    driver.implicitly_wait(10) # wait time
    res = driver.page_source
    obj = bs(res,'html.parser')

    source=driver.page_source
    bs_obj = bs(source, "html.parser")

    for i in bs_obj.findAll('a',{'data-hook':'review-title'}):
        titles.append(i.get_text().strip())
    for n in bs_obj.findAll('span',{'data-hook':'review-date'}):
        nn = ''.join(n.get_text().split(' ')[-3:])
        date = datetime.strptime(nn, '%d%B%Y').date()
        dates.append(date)
    for a in bs_obj.findAll('span',{'data-hook':'review-body'}):
        contents.append(a.get_text().strip())
    for u in bs_obj.findAll('i', {'data-hook':'review-star-rating'}):
        stars.append(int(u.get_text()[0]))
    print('number of Scrped reviews : ', len(titles))
    #driver.find_element(by=By.XPATH,value = '//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a').click()

df2 = pd.DataFrame({'Dates':dates,'Ratings':stars, "Titles":titles, "Bodys":contents})
```
